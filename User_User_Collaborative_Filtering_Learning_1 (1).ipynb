### Import Python Libraries

```python
import pandas as pd
import numpy as np
import scipy.stats
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
```

### Read Data

```python
#Read in the data
ratings = pd.read_csv('Documents/ratings.csv')
ratings.head()
```

```python
#Get dataset info
ratings.info()
```

```python
# Number of users
print('The ratings dataset has', ratings['userId'].nunique(), 'unique users')

#Number of movies
print('The ratings dataset has', ratings['movieId'].nunique(), 'unique movies')

#Number of ratings
print('The ratings dataset has', ratings['rating'].nunique(), 'unique ratings')

#List of ratings
print('The unique ratings are', sorted(ratings['rating'].unique()))
```

```python
#Read in the movie data
movies = pd.read_csv('Documents/movies.csv')
movies.head()
```

```python
movies.info()
```

```python
#Merge ratings and movies dataset 
df = pd.merge(ratings, movies, on='movieId', how='inner')
df.head()
```

### Exploratory Data Analysis (EDA)

```python
#Aggregate by Movie
agg_ratings = df.groupby('title').agg(mean_rating = ('rating', 'mean'), number_of_ratings = ('rating','count')).reset_index()
agg_ratings.head()

#Keep the movies with over 100 ratings
agg_rating_GT100 = agg_ratings[agg_ratings['number_of_ratings'] > 100]
agg_rating_GT100.info()
```

```python
# Check Popular movies
agg_rating_GT100.sort_values(by='number_of_ratings', ascending=False).head()
```

```python
#Correlation between the average rating and number of ratings visulization
sns.jointplot(x='mean_rating', y='number_of_ratings', data=agg_rating_GT100)
```

```python
#Merge data to only keep the movies with more than 100 ratings
df_GT100 = pd.merge(df, agg_rating_GT100[['title']], on='title', how='inner')
df_GT100.info()
```

```python
# Number of users
print('The ratings dataset has', df_GT100['userId'].nunique(), 'unique users')

#Number of movies
print('The ratings dataset has', df_GT100['movieId'].nunique(), 'unique movies')

#Number of ratings
print('The ratings dataset has', df_GT100['rating'].nunique(), 'unique ratings')

#List of ratings
print('The unique ratings are', sorted(df_GT100['rating'].unique()))
```

### Create User-Movie Matrix

```python
#Transform the dataset into a matrix format with the rows being users and the columns being movies. The value of the matrix
# is the user rating of the movie if there is a rating. Otherwise, it shows 'NaN'
matrix = df_GT100.pivot_table(index='userId', columns='title', values='rating')
matrix.head()
```

### Data Normalization

```python
#Since some people tend to give a higher rating than others, normalize the ranting. 
matrix_norm = matrix.subtract(matrix.mean(axis=1), axis='rows')
matrix_norm.head()
```

### Identify Similar Users

```python
#User similarity using Pearsen correlation
user_similarity_pearson = matrix_norm.T.corr(method='pearson')
user_similarity_pearson.head()
```

```python
#User similarity using cosine similarity 
user_similarity_cosine = cosine_similarity(matrix_norm.fillna(0))
user_similarity_cosine
```

```python
#Pick a user ID
picked_userid = 1

#Removed picked user ID from the candidate list
user_similarity_pearson.drop(index=picked_userid, inplace=True)

#Take a look at the data
user_similarity_pearson.head()
```

```python
# Number of similar users
n = 10

#User simialrity threashold
user_similarity_threashold = 0.3

#Get top n similar users
similar_users = user_similarity_pearson[user_similarity_pearson[picked_userid]>user_similarity_threashold][picked_userid].sort_values(ascending=False)[:n]

#print out the top n similar users
print(f'The similar users for user {picked_userid} are', similar_users)
```

### Narrow Down Item Pool

```python
# Movies that the target user has watched
picked_userid_watched = matrix_norm[matrix_norm.index == picked_userid].dropna(axis = 1, how = 'all')
picked_userid_watched
```

```python
#Movies that simialr users watched. Remove movies that none of the simialr users have watched
similar_user_movies = matrix_norm[matrix_norm.index.isin(similar_users.index)].dropna(axis = 1, how='all')
similar_user_movies
```

```python
#Remove the watched movies from the movie list
similar_user_movies.drop(picked_userid_watched.columns,axis=1,inplace=True, errors='ignore')
similar_user_movies
```

### Recommended Items

```python
# A dictionary to store item scores
item_score = {}
# Loop through items
for i in similar_user_movies.columns:
  # Get the ratings for movie i
  movie_rating = similar_user_movies[i]
  # Create a variable to store the score
  total = 0
  # Create a variable to store the number of scores
  count = 0
  # Loop through similar users
  for u in similar_users.index:
    # If the movie has rating
    if pd.isna(movie_rating[u]) == False:
      # Score is the sum of user similarity score multiply by the movie rating
      score = similar_users[u] * movie_rating[u]
      # Add the score to the total score for the movie so far
      total += score
      # Add 1 to the count
      count +=1
  # Get the average score for the item
  item_score[i] = total / count
# Convert dictionary to pandas dataframe
item_score = pd.DataFrame(item_score.items(), columns=['movie', 'movie_score'])
    
# Sort the movies by score
ranked_item_score = item_score.sort_values(by='movie_score', ascending=False)
# Select top m movies
m = 10
ranked_item_score.head(m)
```
